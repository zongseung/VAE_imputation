```python
import numpy as np
import pandas as pd
import torch
import random
from torch import nn
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import ParameterSampler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# ‚úÖ ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import random

# ‚úÖ CUDA ÏÑ§Ï†ï
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"‚úÖ Using device: {device}")
```

```python
# ‚úÖ Beta-VAE Î™®Îç∏ Ï†ïÏùò
class VAE(nn.Module):
    def __init__(self, input_dim, latent_dim, beta=0.5):
        super(VAE, self).__init__()
        self.beta = beta

        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU()
        )

        self.mu = nn.Linear(64, latent_dim)
        self.log_var = nn.Linear(64, latent_dim)

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim),
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.encoder(x)
        return self.mu(h), self.log_var(h)

    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x):
        mu, log_var = self.encode(x)
        z = self.reparameterize(mu, log_var)
        return self.decode(z), mu, log_var
```

```python
def generate_data():
    """‚úÖ ÎûúÎç§ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± Î∞è Í≤∞Ï∏°Í∞í Ï∂îÍ∞Ä"""
    data = np.random.rand(1000, 10)
    data[np.random.rand(*data.shape) < 0.2] = np.nan  # 20% Í≤∞Ï∏°Í∞í Ï∂îÍ∞Ä
    return pd.DataFrame(data, columns=[f'feature_{i}' for i in range(data.shape[1])])


def preprocess_data(df):
    """‚úÖ Í≤∞Ï∏°Í∞íÏùÑ Ï†ïÍ∑úÎ∂ÑÌè¨Î°ú Ï¥àÍ∏∞Ìôî Î∞è Tensor Î≥ÄÌôò"""
    data = df.values
    mask = ~np.isnan(data)

    # ‚úÖ Í≤∞Ï∏°Í∞íÏùÑ ÌèâÍ∑†=0.5, ÌëúÏ§ÄÌé∏Ï∞®=0.2Ïùò Ï†ïÍ∑úÎ∂ÑÌè¨Î°ú Ï¥àÍ∏∞Ìôî
    data[np.isnan(data)] = np.random.normal(loc=0.5, scale=0.2, size=np.isnan(data).sum())
    data = np.clip(data, 0, 1)  # ‚úÖ 0~1 Î≤îÏúÑÎ°ú Ï†úÌïú

    # ‚úÖ PyTorch Tensor Î≥ÄÌôò
    data_tensor = torch.FloatTensor(data).to(device)
    mask_tensor = torch.BoolTensor(mask).to(device)

    return data_tensor, mask_tensor
```

```python
def loss_function(x_hat, x, mu, log_var, mask, beta=0.5):
    """‚úÖ Ïû¨Íµ¨ÏÑ± ÏÜêÏã§Í≥º KL divergence Ìè¨Ìï®"""
    reconstruction_loss = ((x_hat - x) * mask) ** 2
    reconstruction_loss = reconstruction_loss.sum() / mask.sum()

    kl_loss = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())

    return reconstruction_loss + beta * kl_loss
```

```python
def train_vae(model, data_tensor, mask_tensor, learning_rate=0.001, epochs=100, batch_size=64):
    """‚úÖ VAE Î™®Îç∏ ÌïôÏäµ"""
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    model.train()

    for epoch in range(epochs):
        optimizer.zero_grad()

        # ‚úÖ ÏàúÏ†ÑÌåå
        reconstructed, mu, log_var = model(data_tensor)
        loss = loss_function(reconstructed, data_tensor, mu, log_var, mask_tensor, beta=model.beta)

        # ‚úÖ Ïó≠Ï†ÑÌåå Î∞è ÏóÖÎç∞Ïù¥Ìä∏
        loss.backward()
        optimizer.step()

        if epoch % 10 == 0:
            print(f"‚úÖ Epoch [{epoch}/{epochs}], Loss: {loss.item():.4f}")
```

```python
def impute_missing_values(model, data_tensor, mask_tensor, iterations=10):
    """‚úÖ Í≤∞Ï∏°Í∞í Î≥µÏõê (Noise Ï∂îÍ∞Ä)"""
    model.eval()
    with torch.no_grad():
        for _ in range(iterations):
            reconstructed, _, _ = model(data_tensor)
            reconstructed += torch.normal(mean=0, std=0.1, size=reconstructed.shape).to(device)  # ‚úÖ Noise Ï∂îÍ∞Ä
            reconstructed = torch.clamp(reconstructed, 0, 1)  # ‚úÖ 0~1 Î≤îÏúÑÎ°ú Ï†úÌïú
            data_tensor[~mask_tensor] = reconstructed[~mask_tensor]  # ‚úÖ Í≤∞Ï∏°Í∞í Î∂ÄÎ∂ÑÎßå Î≥µÏõê
    return data_tensor.cpu().numpy()
```

```python
def random_search(df, param_list):
    """‚úÖ ÎûúÎç§ ÏÑúÏπòÎ•º ÌÜµÌï¥ ÏµúÏ†ÅÏùò ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ÌÉêÏÉâ"""
    results = []
    best_score = float('inf')
    best_params = None
    best_imputed_data = None

    data_tensor, mask_tensor = preprocess_data(df)

    for i, params in enumerate(param_list):
        print(f"\nüîé Random Search Iteration {i+1}/{len(param_list)}")
        print(params)

        try:
            model = VAE(input_dim=data_tensor.shape[1],
                        latent_dim=params['latent_dim'],
                        beta=params['beta']).to(device)

            train_vae(model, data_tensor.clone(), mask_tensor,
                      learning_rate=params['learning_rate'],
                      epochs=params['epochs'],
                      batch_size=params['batch_size'])

            imputed_data = impute_missing_values(model, data_tensor.clone(), mask_tensor,
                                                 iterations=params['iterations'])

            score = np.nanmean((df.values - imputed_data) ** 2) ** 0.5  # ‚úÖ RMSE
            print(f"‚úÖ RMSE: {score:.4f}")

            if score < best_score and not np.isnan(score):
                best_score = score
                best_params = params
                best_imputed_data = imputed_data
        except Exception as e:
            print(f"‚ùå Error during training: {e}")

        results.append({'params': params, 'score': score if 'score' in locals() else np.nan})

    return results, best_params, best_imputed_data
```

```python
def plot_results(df, imputed_data):
    """‚úÖ ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÏôÄ Î≥µÏõê Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî"""
    plt.figure(figsize=(8, 6))
    plt.hist(df.values.flatten(), bins=50, alpha=0.5, label='Original Data (with NaNs)')
    plt.hist(imputed_data.flatten(), bins=50, alpha=0.5, label='Imputed Data (VAE, Improved)')
    plt.legend()
    plt.title("Original vs. Improved Imputed Data Distribution")
    plt.show()
```

```python
from itertools import product

if __name__ == "__main__":
    # ‚úÖ Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
    df = generate_data()

    # ‚úÖ Grid Search ÌååÎùºÎØ∏ÌÑ∞ Î™©Î°ù
    param_grid = {
        'learning_rate': [0.0005, 0.001],
        'latent_dim': [16, 32],
        'iterations': [10, 20],
        'epochs': [100, 150, 200],
        'batch_size': [32, 64],
        'beta': [0.1, 0.5, 1.0]
    }

    # ‚úÖ Î™®Îì† Ï°∞Ìï©ÏùÑ ÏÉùÏÑ±
    param_list = [dict(zip(param_grid.keys(), values)) for values in product(*param_grid.values())]

    # ‚úÖ Grid Search Ïã§Ìñâ
    results, best_params, best_imputed_data = random_search(df, param_list)

    # ‚úÖ Í≤∞Í≥º ÏãúÍ∞ÅÌôî
    if best_imputed_data is not None:
        plot_results(df, best_imputed_data)
        print("\n‚úÖ Best Hyperparameters:")
        print(best_params)
    else:
        print("\n‚ùå No valid results found.")
```

```python
# ‚úÖ RMSE Ïô∏Ïóê MAE Î∞è MSE Ï∂îÍ∞Ä ÌèâÍ∞Ä
def evaluate_performance(original, imputed):
    mask = ~np.isnan(original)
    mae = np.mean(np.abs(original[mask] - imputed[mask]))
    mse = np.mean((original[mask] - imputed[mask])**2)
    rmse = np.sqrt(mse)
    print(f"üìà MAE: {mae:.4f}, MSE: {mse:.4f}, RMSE: {rmse:.4f}")

# ‚úÖ ÌèâÍ∞Ä Ïã§Ìñâ
evaluate_performance(df.values, best_imputed_data)
```
